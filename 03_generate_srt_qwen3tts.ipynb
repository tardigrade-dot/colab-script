#### Qwen3-ForcedAligner-0.6B


```python
import torch
from qwen_asr import Qwen3ForcedAligner

model = Qwen3ForcedAligner.from_pretrained(
    "/Volumes/sw/pretrained_models/Qwen3-ForcedAligner-0.6B",
    dtype=torch.bfloat16,
    device_map="mps",
    # attn_implementation="flash_attention_2",
)

def forced_aligner(audio_path, text_path):

    with open(text_path, 'r', encoding='utf-8') as f:
        text = f.read().strip()

    results = model.align(
        audio=audio_path,
        text=text,
        language="Chinese",
    )

    print(results[0])
    print(results[0][0].text, results[0][0].start_time, results[0][0].end_time)

forced_aligner(
    "/Volumes/sw/tts_result/sulianjianshi/sulianjianshi-7_3.wav", 
    "/Volumes/sw/tts_result/sulianjianshi/sulianjianshi-7_3.txt"
    )
```

```python
import torch
from qwen_asr import Qwen3ASRModel

device = "cpu" #cuda:0
model = Qwen3ASRModel.from_pretrained(
    "/Volumes/sw/pretrained_models/Qwen3-ASR-1.7B",
    dtype=torch.bfloat16,
    device_map=device,
    # attn_implementation="flash_attention_2",
    max_inference_batch_size=32, # Batch size limit for inference. -1 means unlimited. Smaller values can help avoid OOM.
    max_new_tokens=256, # Maximum number of tokens to generate. Set a larger value for long audio input.
)

results = model.transcribe(
    audio="/Users/larry/github.com/tardigrade-dot/colab-script2/data_src/tianchaoyaoyuan2-7_3.wav",
    language="Chinese", # set "English" to force the language
)

print(results[0].language)
print(results[0].text)
```
