```python
config = """
lines: AUTO
limit: 3
streamers:
  "/content/drive/MyDrive/zzqlyzgsh2/*.mp4":
    copyright: 1 ### 1为自制，2为转载
    tid: 36 # 投稿分区
    # cover: "" # 视频封面
    title: 测试标题
    desc: "https://book.douban.com/subject/1326771/"
    tags:
      - "有声书"
    dtime: ~
    open_subtitle: false

"""
```

```python
with open('config.yml', 'w') as f:
    f.write(config)
```

```python
!biliup --user-cookie \
  /content/drive/MyDrive/biliup_cookies.json \
  upload \
  -c /content/drive/MyDrive/bili_config.yml
```
